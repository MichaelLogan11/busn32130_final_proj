```python
import pandas as pd
import yfinance as yf
import re
```

```python
# Load in data and create "tickers" list
tickers = ['GBTC', 'ETHE', 'SPY', 'BND', 'GLD', 'VNQ']

# Compile all data in "equity_data" dataframe
equity_data = pd.DataFrame()
start_date = "2000-01-01"
end_date = "2025-01-01"

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date, group_by='ticker')

    if isinstance(data.columns, pd.MultiIndex):
        df = data[ticker].copy()
    else:
        df = data.copy()

    df = df.reset_index()
    df['Ticker'] = ticker
    df = df[['Date', 'Ticker', 'Open', 'High', 'Low', 'Close', 'Volume']]

    equity_data = pd.concat([equity_data, df], ignore_index=True)

equity_data.columns.name = None

# Sort by Ticker and Date to ensure proper order
equity_data = equity_data.sort_values(['Ticker', 'Date']).reset_index(drop=True)

# Calculate Net Change and % Change in Close for each Ticker
equity_data['Net Change'] = equity_data.groupby('Ticker')['Close'].diff()
equity_data['% Change'] = equity_data.groupby('Ticker')['Close'].pct_change() * 100

# Export csv
equity_data.to_csv("equity_data.csv", index=False)
```
